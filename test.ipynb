```python
conf = sp.parse_params('byi_second_bs_V230324', '15分钟_D1MACD12#26#9回抽零轴_BS2辅助V230324_看空_任意_任意_0')
signals_config.append(conf)
```

```python
bars = get_raw_bars('BTCUSDT', '1分钟', sdt= '2020-01-01', edt = '2020-03-01')
```

```python
[{'freq': '15分钟',
  'di': 1,
  'fastperiod': '12',
  'slowperiod': '26',
  'signalperiod': '9',
  'name': 'czsc.signals.byi_second_bs_V230324'}]
sigs = czsc.generate_czsc_signals(bars, signals_config=signals_config, df=True)
sigs
```

```python
sigs.loc[sigs['15分钟_D1MACD12#26#9回抽零轴_BS2辅助V230324'] != '其他_任意_任意_0']
```

```python
def set_value(df, data_column, result_column):
    df.loc[df['15分钟_D1_表里关系V230102'] == '其他_任意_任意_0', result_column] = 0
    df.loc[df['15分钟_D1_表里关系V230102'] == '向下_底分_任意_0', result_column] = 1
    df.loc[df['15分钟_D1_表里关系V230102'] == '向下_延伸_任意_0', result_column] = 2
    df.loc[df['15分钟_D1_表里关系V230102'] == '向上_顶分_任意_0', result_column] = 3
    df.loc[df['15分钟_D1_表里关系V230102'] == '向上_延伸_任意_0', result_column] = 4
    return df
```

```python
sigs = set_value(sigs, '1分钟_D1_表里关系V230102', 'result1')
sigs = set_value(sigs, '3分钟_D1_表里关系V230102', 'result3')
sigs = set_value(sigs, '5分钟_D1_表里关系V230102', 'result5')
sigs = set_value(sigs, '15分钟_D1_表里关系V230102', 'result15')
sigs = set_value(sigs, '30分钟_D1_表里关系V230102', 'result30')
sigs = set_value(sigs, '60分钟_D1_表里关系V230102', 'result60')
sigs = set_value(sigs, '120分钟_D1_表里关系V230102', 'result120')
```
